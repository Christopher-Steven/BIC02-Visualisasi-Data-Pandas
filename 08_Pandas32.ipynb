## Plotting dari suatu pandas Data Frame 

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

### Data Frame

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows,n_cols)),
                 columns = cols)

df.head()
```

### Line Plot

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A','B']].plot(kind='line')
```

> `Plot(kind='line')`
>> - Parameter pertama digunakan untuk membuat plotingan bentuknya garis atau line.
>> - Plot ini akan menampilkan semua kolom pada DataFrame
>> - Sumbu x merepresentasikan sebagai baris atau nilai index
>> - Sumbu Y merepresentasikan nilai yang tertampung pada DataFrame
>> - Membuat spesifik kolom maka memerlukan fungsi `[val_cols1,val_cols2,n].plot(kind='line')`.

### Bar Plot

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A','B']].plot(kind='bar')
```

```python
df[['A','B']].head().plot(kind='bar')
```

```python
df[['A','B']].head().plot(kind='barh')
```

> `plot(kind='bar')`
>> - Parameter kesatu yang digunakan ntuk membuat plotingan berbentuk bar yaitu `kind='bar'` dan untuk membuat bar dalam bentuk horizontal maka menggunakan `kind='barh'`.
>> - Sumbu x merepresentasikan sebagai baris atau nilai index
>> - Sumbu Y merepresentasikan nilai yang tertampung pada DataFrame
>> - untuk menampilkan 5 data maka menggunakan head().
>> - Membuat spesifik kolom maka memerlukan fungsi `[val_cols1,val_cols2,n].plot(kind='line')`.

### Area Plot

### Box Plot

```python
df.head()
df.plot(kind='box')
```

> `plot(kind='box')`
>> - Parameter kesatu yang digunakan ntuk membuat plotingan berbentuk box yaitu `kind='box'`.
>> - untuk sumbu x merepresentasikan sebagai nama kolomnya bukan index.
>> - untuk sumbu y merepresentasikan rentan nilai dari DataFrame.

```python
df[['A','B']].plot(kind='hist')
```

### Scatter plot

```python
df.head()
```

```python
df.plot(x='A', y='B', kind= 'scatter')
```

> `plot(x='val_cols', y='val_cols', kind='scatter')`
>> - parameter pertama digunakan untuk nenetukan sumbu x yang bernilai nama kolom
>> - parameter kedua digunakan untuk nenetukan sumbu y yang bernilai nama kolom
>> - parameter ketiga menggunakan `kind='scatter'` digunakan untuk membuat scatter plot. namun untuk sumbu x dan y tidak memiliki kolerasi.
