#### membuat data frame

```python
df = pd.read_csv('./titanicfull.csv')
df.columns = ['Pclass','Surivival status', 'full Name', 'Sex ', 
              ' Age', 'sib SP','Parch','Ticket','Fare','Cabin',
             'Embarked']
df_backup = df.copy(deep = True)
df.head()
```

- untuk mengread data dari file lain menggunakan `read_filetype('./folder(optional)/file.typefile)`
- `df.head()` digunakan untuk menampilkan 5 data pertama
- `df.copy(deep = True)` digunakan untuk mengcopy data awal

#### Menggunakan lowercase untuk nama kolom dan mengganti spasi  dengan tanda `_`

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.head()
```

- `str.replace('current text','new text')` mengganti tulisan yang mau di ubah
- `str.lower()` digunakan untuk mengganti huruf besar menjadi huruf kecil


#### memangkas spasi pada nama kolom

```python
df = df_backup.copy(deep= True)
df.columns = df.columns.str.lower().str.strip().str.replace(' ','_')
df.head()
```

- untuk cara kerja kodingan `df.columns = df.columns.str.lower().str.strip().str.replace(' ','_')` bagian ini dilakuakan secara bertahap dan berurutan.
- `df.columns` pada bagian ini saya akan melakukan akan modifikasi pada data frame ini.
- `str.lower()` saya akan melakukan lower case dulu (mengubah huruf besar menjadi huruf kecil).
- `str.strip()` bagian ini akan memotong bagian spasi disisi kiri dan kanan.
- `str.replace('old string','new string')` bagian ini akan mengubah bagian kata yang dulu (parameter 1) dengan yang sekarang (parameter 2)
